# Final Project  6: Corporate Bond Liquidity

## Multicurve Bootstrap

```python
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
from datetime import datetime

import utility as ut
import dates_manipulation as dm
import mainFunctions as mf
import auxiliars as aux
```

##### Import data from Excel file

```python
dataFileName = "./data/curves20150910_project.xlsx"
[rates, dates, normal_vols] = ut.read_file_excel(dataFileName)
```

##### Clean dates

```python
dates['ois'] = dm.first_date_MF( dates['ois'], dm.eurCalendar() )
dates['fra_start'] = dm.first_date_MF( dates['fra_start'], dm.eurCalendar() )
dates['fra_end'] = dm.first_date_MF( dates['fra_end'], dm.eurCalendar() )
dates['euribor'] = dm.first_date_MF( dates['euribor'], dm.eurCalendar() )
dates['swaps'] = dm.first_date_MF( dates['swaps'], dm.eurCalendar() )
```

### Bootstrap OIS

We use a standard bootstrap technique, applying different relations according to OIS maturity (greater or smaller than 1y)

```python
#Perform the bootstrap of OIS curve
[dates_EONIA, discounts_EONIA] = mf.bootstrap_OIS(dates,rates)
```

### Bootstrap pseudo-curve

We use the "Mr. Crab's Bootstrap" technique on Euribor6m using 6m depos, 6m FRAs and swaps vs Euribor6m  

```python
#Perfrom the bootstrap of pseudo-discount curve
[dates_pseudo, discount_pseudo] = mf.bootstrap_pseudo(dates,rates, dates_EONIA, discounts_EONIA)
```

##### Plot of discount factors curves ( EONIA curve in red, pseudo curve in blue)

```python
myticks=list(map(str,list(map(lambda x: datetime.fromordinal(int(x)).strftime("%d-%b-%y"),list(dates_EONIA)))))
plt.xticks([dates_EONIA[i] for i in [0,7,9,11,13,15,18]],[myticks[i] for i in [0,7,9,11,13,15,18]],rotation=45)
plt.plot(list(dates_EONIA), list(discounts_EONIA),'r', dates_pseudo, discount_pseudo, 'b')
plt.ylabel('discount_factors')
plt.legend(['EONIA DF curve', 'Euribor DF curve'])
plt.title("Discount Curves")
plt.grid()
plt.show()
```

##### Plot of zero-rates curves

```python
# computing zero_rates
day_count_act_365 = 3
zr_1 = np.zeros(len(discounts_EONIA)-1)
for i in range(1,len(discounts_EONIA)):
    yf = dm.yearfrac([dates['settle']],[dates_EONIA[i]], day_count_act_365)
    zr_1[i-1] = -np.log(discounts_EONIA[i])/ yf
    
zr_2 = np.zeros(len(discount_pseudo)-1)
for i in range(1,len(discount_pseudo)):
    yf = dm.yearfrac([dates['settle']],[dates_pseudo[i]], day_count_act_365)
    zr_2[i-1] = -np.log(discount_pseudo[i])/ yf
    
# plot
myticks=list(map(str,list(map(lambda x: datetime.fromordinal(int(x)).strftime("%d-%b-%y"),list(dates_EONIA)))))
plt.xticks([dates_EONIA[i] for i in [0,7,9,11,13,15,18]],[myticks[i] for i in [0,7,9,11,13,15,18]], rotation=45)
plt.plot(dates_EONIA[1::], zr_1, 'r', dates_pseudo[1::], zr_2, 'b')
plt.ylabel('zero_rates')
plt.legend(['EONIA z-rates', 'Euribor z-rates'])
plt.title("Rates Curves")
plt.grid()
plt.show()
```

## Calibration of the model

We calibrate model parameters minimizing the $L^2$ distance between model and market prices functions of ATM CS-receiver "diagonal" swaptions

```python
# Calibrate the model
[RS_calibrated_prices, RS_mkt_prices, calibrated_params]=mf.calibrate_model(normal_vols, dates_EONIA,\
                                                          discounts_EONIA, dates_pseudo, discount_pseudo)
```

```python
# Plot
plt.plot( normal_vols['expiry'], RS_mkt_prices, 'rs-', normal_vols['expiry'], RS_calibrated_prices, 'bd:')
plt.ylabel('prices')
plt.xlabel('tenors')
plt.legend(['Mkt Prices', 'Model Prices'])
plt.title('Mkt vs Model prices')
plt.grid()
plt.show()
```

## Liquidity: BNPP & Santander cases

##### Define time to liquidate

```python
#time to liquidate: 2 cases
date_two_weeks=dm.dateMoveVec_MF([dates['settle']], 'w', 2, dm.eurCalendar())
date_two_months=dm.dateMoveVec_MF([dates['settle']], 'm', 2, dm.eurCalendar())
```

##### Import data and compute dirty prices

```python
[BNPP, Santander]=aux.buildStruct()
```

```python
dirty_prices_BNPP=ut.dirty_from_clean(BNPP["dates"], BNPP["coupons"], BNPP["clean_prices"], dates["settle"])
dirty_prices_Santander=ut.dirty_from_clean(Santander["dates"], Santander["coupons"],\
                                           Santander["clean_prices"], dates["settle"])
```

### Compute liquidity bounds in case of 2w and 2m time to liquidate

```python
#BNPP

#two weeks
[upper_bound_BNPP_2w, lower_bound_BNPP_2w]=mf.liquidity_spread_bounds(BNPP,\
                                        BNPP["surv_probs"][0], BNPP["dates"], calibrated_params,\
 dates["settle"], date_two_weeks, dirty_prices_BNPP, discounts_EONIA, dates_EONIA)
error_BNPP_2w=np.abs(np.array(upper_bound_BNPP_2w)-np.array(lower_bound_BNPP_2w))
#two months
[upper_bound_BNPP_2m, lower_bound_BNPP_2m]=mf.liquidity_spread_bounds(BNPP,\
                                        BNPP["surv_probs"][1], BNPP["dates"], calibrated_params,\
 dates["settle"], date_two_months, dirty_prices_BNPP, discounts_EONIA, dates_EONIA)
error_BNPP_2m=np.abs(np.array(upper_bound_BNPP_2m)-np.array(lower_bound_BNPP_2m))
```

```python
#Santander

#two weeks
[upper_bound_Santander_2w, lower_bound_Santander_2w]=mf.liquidity_spread_bounds(Santander,\
                                    Santander["surv_probs"][0], Santander["dates"], calibrated_params,\
 dates["settle"], date_two_weeks, dirty_prices_Santander, discounts_EONIA, dates_EONIA)
error_Santander_2w=np.abs(np.array(upper_bound_Santander_2w)-np.array(lower_bound_Santander_2w))
#two months
[upper_bound_Santander_2m, lower_bound_Santander_2m]=mf.liquidity_spread_bounds(Santander,\
                                    Santander["surv_probs"][1], Santander["dates"], calibrated_params,\
 dates["settle"], date_two_months, dirty_prices_Santander, discounts_EONIA, dates_EONIA)
error_Santander_2m=np.abs(np.array(upper_bound_Santander_2m)-np.array(lower_bound_Santander_2m))      
```

##### Plot the max errors in liquidity price, i.e. the spread between upper and lower bound

*BNPP liquidity price error*

```python
ut.do_plot_error(dates["settle"],BNPP["dates"],error_BNPP_2w,error_BNPP_2m,'BNPP')
```

*Santander liquidity price error*

```python
ut.do_plot_error(dates["settle"],Santander["dates"],error_Santander_2w,error_Santander_2m,'Santander')
```

### Compute bond yields

*BNPP bond yields*

```python
#BNPP

#two weeks
[liq_yield_BNPP_2w, bond_yield_BNPP_2w]=mf.bond_yield(date_two_weeks, BNPP["dates"],\
                                        dates["settle"],dirty_prices_BNPP,upper_bound_BNPP_2w, BNPP["coupons"])

[liq_yield_BNPP_2m, bond_yield_BNPP_2m]=mf.bond_yield(date_two_months, BNPP["dates"],\
                                        dates["settle"],dirty_prices_BNPP,upper_bound_BNPP_2m, BNPP["coupons"])
```

*Santander bond yields*

```python
#Santander

#two months
[liq_yield_Santander_2w, bond_yield_Santander_2w]=mf.bond_yield(date_two_weeks, Santander["dates"],\
                                        dates["settle"],dirty_prices_Santander,upper_bound_Santander_2w, Santander["coupons"])

[liq_yield_Santander_2m, bond_yield_Santander_2m]=mf.bond_yield(date_two_months, Santander["dates"],\
                                        dates["settle"],dirty_prices_Santander,upper_bound_Santander_2m, Santander["coupons"])
```

##### Plot: bond yields vs maturity

*BNPP bond yields*

```python
ut.do_plot_yield(dates["settle"],BNPP["dates"],\
                 bond_yield_BNPP_2w,liq_yield_BNPP_2w,liq_yield_BNPP_2m,'BNPP')
```

*Santander bond yields*

```python
ut.do_plot_yield(dates["settle"],Santander["dates"],\
                 bond_yield_Santander_2w,liq_yield_Santander_2w,liq_yield_Santander_2m,'Santander')
```
